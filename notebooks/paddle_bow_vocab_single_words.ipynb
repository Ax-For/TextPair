```python
import os
```

```python
VOCAB_FILE = '../data/paddle_models/sim_net/data/term2id.dict'
```

```python
def load_vocab(file_path):
    """
    load the given vocabulary
    """
    vocab = {}
    if not os.path.isfile(file_path):
        raise ValueError("vocabulary dose not exist under %s" % file_path)
    with open(file_path, 'r', encoding="utf-8") as f:
        for line in f:
            items = line.strip('\n').split("\t")
            if items[0] not in vocab:
                vocab[items[0]] = int(items[1])
    vocab["<unk>"] = 0
    return vocab
```

```python
vocab = load_vocab(VOCAB_FILE)
```

```python
single_words = { k:v for k, v in vocab.items() if len(k) == 1}
```
