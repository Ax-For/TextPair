```python
import os
import pickle

import numpy as np
import paddle.fluid as fluid
```

```python
DATA_PATH = '../data'
PADDLE_MODELS_PATH = os.path.join(DATA_PATH, 'paddle_models')
SIM_NET_PATH = os.path.join(PADDLE_MODELS_PATH, 'sim_net')
MODEL_PATH = os.path.join(SIM_NET_PATH, 'model_files/simnet_bow_pairwise_pretrained_model/')
VOCAB_PATH = os.path.join(SIM_NET_PATH, 'data/term2id.dict')

WORD_VEC_DIR = os.path.join(SIM_NET_PATH, 'word_vec')
if not os.path.exists(WORD_VEC_DIR):
    os.makedirs(WORD_VEC_DIR)
    
WORD_VEC_PKL = os.path.join(WORD_VEC_DIR, 'paddle_bow_wv.pkl')

EMBEDDING_NAME = 'emb'
```

```python
place = fluid.CPUPlace()
executor = fluid.Executor(place)

program, feed_var_names, fetch_targets = fluid.io.load_inference_model(MODEL_PATH, executor)
emb = fluid.global_scope().find_var(EMBEDDING_NAME).get_tensor()
```

```python
def load_vocab(file_path):
    """
    load the given vocabulary
    """
    vocab = {}
    if not os.path.isfile(file_path):
        raise ValueError("vocabulary dose not exist under %s" % file_path)
    with open(file_path, 'r', encoding="utf-8") as f:
        for line in f:
            items = line.strip('\n').split("\t")
            if items[0] not in vocab:
                vocab[items[0]] = int(items[1])
    vocab["<unk>"] = 0
    return vocab
```

```python
vocab = load_vocab(VOCAB_PATH)
```
